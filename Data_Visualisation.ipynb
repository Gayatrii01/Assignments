# DATA VISUALIZATION ASSIGNMENT

#### Problem Statement:
This assignment aims to perform data visualization on two datasets — the Iris dataset and the Diabetes dataset — using Matplotlib and Seaborn. The goal is to explore and interpret data visually, uncovering relationships, trends, and patterns that contribute to deeper insights and understanding.

#### Dataset 1: Iris Dataset:

Source: https://www.kaggle.com/datasets/uciml/iris/data

Description: The Iris dataset contains measurements of 150 iris flowers from three different species: Iris-setosa, Iris-versicolor, and Iris-virginica. It includes four features:

Sepal Length

Sepal Width

Petal Length

Petal Width

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Load and Inspect the Datasets

```python
# Load the Iris dataset
iris_df = pd.read_csv('Iris1.csv')
```

```python
# Display basic info
print("Shape of dataset:", iris_df.shape)
print("\nColumn names:", iris_df.columns.tolist())
```

```python
iris_df.head()
```

### 2. Exploratory Data Visualization – Iris Dataset


```python
# Drop the 'Id' column as it's not useful for analysis
iris_df = iris_df.drop('Id', axis=1)
```

```python
iris_df.head()
```

### Question 1: What is the distribution of sepal lengths and widths among the Iris species?

```python
#  Set the visual style
sns.set(style="whitegrid")
```

```python
# Create subplots and Set up the plotting area
plt.figure(figsize=(14, 6))
```

```python
# Plot distribution of Sepal Length by Species
plt.subplot(1, 2, 1)
sns.histplot(data=iris_df, x='SepalLengthCm', hue='Species', kde=True, palette='Set2')
plt.title('Distribution of Sepal Length by Species')
```

```python
# Plot distribution of Sepal Width by Species
plt.subplot(1, 2, 2)
sns.histplot(data=iris_df, x='SepalWidthCm', hue='Species', kde=True, palette='Set2')
plt.title('Distribution of Sepal Width by Species')
```

```python
plt.tight_layout()
plt.show()
```

#### Interpretation:
Setosa tends to have shorter sepal lengths and wider sepal widths.

Virginica has longer sepals, while Versicolor falls in the middle range.

### Question 2: How does the petal length vary across different Iris species?

```python
# Create a boxplot to show petal length differences by species
plt.figure(figsize=(8, 5))
sns.boxplot(data=iris_df, x='Species', y='PetalLengthCm', palette='Set2')
plt.title('Petal Length Distribution by Species')
plt.show()
```

####  Interpretation:
Setosa has significantly shorter petals.

Clear distinction in petal length between all three species — useful for classification.

### Question 3: Is there a relationship between Sepal Length and Petal Length?

```python
# Scatter plot to visualize the relationship between Sepal Length and Petal Length
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_df, x='SepalLengthCm', y='PetalLengthCm', hue='Species', palette='Set2')
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()
```

#### Interpretation:

A positive correlation is observed — as sepal length increases, petal length tends to increase.

The relationship is especially distinct for Versicolor and Virginica.

Setosa shows low sepal and petal lengths with minimal overlap with the other two species.

### Question 4: Are there any patterns observable in the pairwise relationships between features?

```python
# Pairplot to visualize pairwise relationships between all features
sns.pairplot(iris_df, hue='Species', palette='Set2', corner=True)
plt.suptitle('Pairwise Feature Relationships in the Iris Dataset', y=1.02)
plt.show()
```

#### Interpretation:

Clear separation between Setosa and the other species in almost all feature combinations.
    
Petal length and width are especially good at distinguishing species.

Versicolor and Virginica overlap a bit, but show distinguishable patterns in petal-based features.

### Question 5: Can we distinguish between Iris species based on sepal and petal measurements?

Used a 2D scatter plot combining:

Petal Length vs Petal Width (strong differentiators)

Colored by species

This will help us visually assess separability between species.

```python
# Scatter plot using Petal Length and Petal Width to distinguish species
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_df, x='PetalLengthCm', y='PetalWidthCm', hue='Species', palette='Set2')
plt.title('Species Separation Using Petal Measurements')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```

#### Interpretation:

Setosa is clearly separable from the other two species using petal measurements.

Versicolor and Virginica overlap slightly but still form distinct clusters.
    
Petal length and width are powerful features for classifying Iris species.

## Dataset 2: Diabetes Dataset

#### Source: https://www.kaggle.com/datasets/mathchi/diabetes-data-set/data

#### About dataset: 

This dataset consists of various health-related variables and a target variable indicating diabetes progression. It includes features such as:

Age

BMI (Body Mass Index)

Blood Pressure

Glucose

Insulin

DiabetesPedigreeFunction

Skin Thickness

Pregnancies

Outcome (0: non-diabetic, 1: diabetic)



### 1. Load and Preview the Diabetes Dataset

```python
# Load the Diabetes dataset
diabetes_df = pd.read_csv('diabetes.csv')
```

```python
# Display basic info
print("Shape of dataset:", diabetes_df.shape)
print("\nColumn names:", diabetes_df.columns.tolist())
```

```python
diabetes_df.head()
```

#### Initial Observations:

Will explore variables such as Age, BMI, BloodPressure, Glucose, and Insulin.

Target variable Outcome indicates diabetic condition.

#### Question 6: What is the distribution of Age and BMI among diabetic patients?

Will use histograms and box plots to visualize how Age and BMI are distributed, separately for diabetic (Outcome = 1) and non-diabetic (Outcome = 0) individuals.

Distribution of Age and BMI :

```python
# Set up visual style
sns.set(style="whitegrid")
```

```python
# Create subplots for Age and BMI
plt.figure(figsize=(14, 6))
```

```python
# Age Distribution
plt.subplot(1, 2, 1)
sns.histplot(data=diabetes_df, x='Age', hue='Outcome', kde=True, palette='Set1')
plt.title('Age Distribution by Diabetes Outcome')
plt.xlabel('Age')
plt.ylabel('Count')
```

```python
# BMI Distribution
plt.subplot(1, 2, 2)
sns.histplot(data=diabetes_df, x='BMI', hue='Outcome', kde=True, palette='Set1')
plt.title('BMI Distribution by Diabetes Outcome')
plt.xlabel('BMI')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

#### Interpretation:
Diabetic patients (Outcome = 1) are more common in higher age and higher BMI ranges.
Non-diabetic individuals are more spread out in younger age groups and lower BMI values.

### Question 7: How do blood pressure levels vary across different age groups?

Will use a box plot to visualize how Blood Pressure varies with Age Groups. First, we’ll bin the age variable into meaningful categories (e.g., 20s, 30s, etc.).

```python
# Create Age Groups (bins of 10 years)
diabetes_df['AgeGroup'] = pd.cut(diabetes_df['Age'], bins=[20, 30, 40, 50, 60, 70, 80], 
                                 labels=['20s', '30s', '40s', '50s', '60s', '70s'])
```

```python
# Boxplot of Blood Pressure by Age Group
plt.figure(figsize=(8, 6))
sns.boxplot(data=diabetes_df, x='AgeGroup', y='BloodPressure', palette='Set3')
plt.title('Blood Pressure Distribution Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Blood Pressure')
plt.show()
```

#### Interpretation:

Blood pressure tends to increase slightly with age, though there’s variation within groups.

The 50s and 60s age groups show slightly higher median blood pressure values.

### Question 8: Is there a correlation between BMI and Glucose levels?

I'll visualize the relationship using a scatter plot with a regression line to see how BMI relates to Glucose. 
A correlation heatmap will also help confirm the strength of the relationship.

```python
# Scatter plot with regression line to observe relationship
plt.figure(figsize=(8, 6))
sns.regplot(data=diabetes_df, x='BMI', y='Glucose', scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Correlation Between BMI and Glucose Levels')
plt.xlabel('BMI')
plt.ylabel('Glucose Level')
plt.show()
```

```python
# Correlation Heatmap for numerical features
plt.figure(figsize=(10, 6))
corr = diabetes_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix - Diabetes Dataset')
plt.show()
```

#### Interpretation:

There’s a moderate positive correlation between BMI and Glucose levels.

Higher BMI may be associated with elevated glucose, which is a key diabetes indicator.

### Question 9: Are there any observable patterns in the relationship between diabetes progression and patient age?

I'll use : 

A count plot to see how diabetes outcome varies across different age groups.

Optionally, a line plot to view average diabetes probability across age.

```python
# Count of diabetic vs non-diabetic in each age group
plt.figure(figsize=(8, 6))
sns.countplot(data=diabetes_df, x='AgeGroup', hue='Outcome', palette='Set2')
plt.title('Diabetes Outcome Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Outcome', labels=['Non-Diabetic', 'Diabetic'])
plt.show()
```

```python
# Optional: Line plot of mean outcome by age
plt.figure(figsize=(10, 5))
age_outcome = diabetes_df.groupby('Age')['Outcome'].mean()
age_outcome.plot(kind='line', color='tomato', marker='o')
plt.title('Average Diabetes Probability by Age')
plt.xlabel('Age')
plt.ylabel('Mean Diabetes Outcome')
plt.grid(True)
plt.show()
```

#### Interpretation:

The proportion of diabetic individuals increases with age, especially in the 40s–60s range.

Younger age groups show a lower prevalence of diabetes.

### Question 10: How do insulin levels vary across different age groups?

I’ll use a box plot to visualize the spread and central tendency of Insulin levels within each age group.

```python
# Box plot for Insulin levels by Age Group
plt.figure(figsize=(8, 6))
sns.boxplot(data=diabetes_df, x='AgeGroup', y='Insulin', palette='pastel')
plt.title('Variation in Insulin Levels Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Insulin Level')
plt.show()
```

#### Interpretation:

Insulin levels vary widely within all age groups, with many outliers.
    
The 30s to 50s age range shows slightly higher median insulin levels.

Variation may indicate irregular insulin secretion patterns in diabetic individuals.
