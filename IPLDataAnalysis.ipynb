## Answers of Data Analysis Questions:

### 1.What was the count of matches played in each season?

```python
matches_per_season = matches_2020['date'].str.slice(0, 4).value_counts().sort_index()
print("Matches played in each season:")
print(matches_per_season)

# Extract the year from the 'date' column and count the number of matches per season (year).
# str.slice(0, 4): Extract the first four characters (year) from the 'date' column.
# value_counts(): Count the occurrences of each unique year.
# sort_index(): Sort the results by the year in ascending order.
```

#### Visualizations of the results by using a bar plot :

```python
plt.figure(figsize=(8, 4))
sns.barplot(x=matches_per_season.index, y=matches_per_season.values, palette="viridis")
plt.title("Matches played in each season", fontsize=16)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Matches played", fontsize=12)
plt.show()
```

### 2. How many runs were scored in each season?

```python
# Extract year from the 'date' column in the 'IPL Matches 2008-2020' file
matches_2020['year'] = pd.to_datetime(matches_2020['date']).dt.year

# Merge datasets on 'id'
merged_data = pd.merge(ball_2020, matches_2020, on='id')

# Group by 'year' and 'id' and calculate runs per match
runs_per_match_per_season = merged_data.groupby(['year', 'id'])['total_runs'].sum()

# Calculate mean runs per match for each season
average_runs_per_match_per_season = runs_per_match_per_season.groupby('year').mean()
print("\nRuns scored in each season:")
print(average_runs_per_match_per_season)
```

#### Visualizations of the results by using a bar plot :

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=average_runs_per_match_per_season.index, y=average_runs_per_match_per_season.values, palette="viridis")
plt.title("Runs scored per match in different seasons", fontsize=16)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Average Runs per Match", fontsize=12)
plt.show()
```

### 3. What were the runs scored per match in different seasons?

```python
# Calculate runs per match for each 'id'
runs_per_match_per_season = merged_data.groupby(['year', 'id'])['total_runs'].sum()

# Calculate mean runs per match for each season
average_runs_per_match_per_season = runs_per_match_per_season.groupby('year').mean()

# Display the results
print("\nRuns per Match for Each 'id' and Season:")
print(runs_per_match_per_season)
```

#### Visualizations of the results by using a boxplot :

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=runs_per_match_per_season.reset_index(), x='year', y='total_runs')
plt.title('Distribution of Runs per Match for Each Season', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Runs per Match', fontsize=12)
plt.show()
```
