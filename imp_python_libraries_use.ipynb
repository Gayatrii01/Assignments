## 3. Pandas DataFrame Manipulation: 
● Read data using pd.read_csv(). 
● Perform operations like selecting columns, filtering rows, and applying functions. 
● Group data using .groupby() and aggregate values. 
● Handle missing values using .fillna(), .dropna().

```python
# Selecting columns and filtering rows
print(df[['total_bill', 'tip']])
```

```python
print(df[df['tip'] > 5])
```

```python
# Grouping and aggregation
grouped = df.groupby('day')['tip'].mean()
print(grouped)
```

```python
# Handle missing values
df = df.dropna()
```

```python
# Convert DataFrame to NumPy array
array_from_df = df[['total_bill', 'tip']].values
```

## 4. Data Preprocessing & Transformation: 
● Normalize numerical columns using NumPy operations. 
● Apply transformations like log() scaling. 
● Convert categorical variables into numerical format using Pandas .get_dummies().

```python
# Normalize a column
df['normalized_total_bill'] = (df['total_bill'] - df['total_bill'].min()) / (df['total_bill'].max() - df['total_bill'].min())
```

```python
# Log transformation
df['log_total_bill'] = np.log(df['total_bill'])
```

```python
# Convert categorical to numeric
df = pd.get_dummies(df, columns=['sex', 'smoker', 'day', 'time'], drop_first=True)
```

## 5. Data Visualization: 
● Generate summary statistics using .describe(). 
● Visualize distributions using Pandas built-in plotting functions. 
● Create histograms, scatter plots, and box plots using Matplotlib. 
● Generate correlation heatmaps and violin plots using Seaborn.

```python
# Summary statistics
print(df.describe())
```

```python
# Matplotlib visualizations
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(df['total_bill'], bins=20, color='skyblue')
plt.title('Histogram of Total Bill')
```

```python
plt.subplot(1, 2, 2)
plt.scatter(df['total_bill'], df['tip'], color='green')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.tight_layout()
plt.show()
```

```python
# Seaborn visualizations
sns.pairplot(df[['total_bill', 'tip', 'size']])
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
# Re-import the original dataset
original_df = pd.read_csv('tips.csv')

sns.boxplot(x='sex', y='tip', data=original_df)
plt.title('Tip Amount by Gender')
plt.show()
```
